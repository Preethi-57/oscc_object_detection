```python
import os
import torch
import kagglehub
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader

# Download dataset
dataset_path = kagglehub.dataset_download("ashenafifasilkebede/dataset")

# Define data transforms
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Load dataset
train_dataset = ImageFolder(root=os.path.join(dataset_path, "train"), transform=transform)
val_dataset = ImageFolder(root=os.path.join(dataset_path, "val"), transform=transform)
test_dataset = ImageFolder(root=os.path.join(dataset_path, "test"), transform=transform)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# Check class names
class_names = train_dataset.classes
print("Classes:", class_names)
```

```python
import torch.nn as nn
import torchvision.models as models

# Function to get CNN models
def get_cnn_model(model_name, num_classes=2):
    if model_name == "resnet50":
        model = models.resnet50(pretrained=True)
        model.fc = nn.Linear(model.fc.in_features, num_classes)

    elif model_name == "efficientnet":
        model = models.efficientnet_b0(pretrained=True)
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, num_classes)

    elif model_name == "densenet":
        model = models.densenet121(pretrained=True)
        model.classifier = nn.Linear(model.classifier.in_features, num_classes)

    return model

# Example usage
cnn_model = get_cnn_model("resnet50")
print(cnn_model)
```

```python
import timm

# Function to get transformer models
def get_transformer_model(model_name, num_classes=2):
    if model_name == "vit":
        model = timm.create_model("vit_base_patch16_224", pretrained=True, num_classes=num_classes)

    elif model_name == "swin":
        model = timm.create_model("swin_base_patch4_window7_224", pretrained=True, num_classes=num_classes)

    return model

# Example usage
vit_model = get_transformer_model("vit")
print(vit_model)
```

```python
# Function to get hybrid models
def get_hybrid_model(model_name, num_classes=2):
    if model_name == "convnext":
        model = timm.create_model("convnext_base", pretrained=True, num_classes=num_classes)

    elif model_name == "cvt":
        model = timm.create_model("cvt-13", pretrained=True, num_classes=num_classes)

    return model

# Example usage
hybrid_model = get_hybrid_model("convnext")
print(hybrid_model)
```

```python
import torch.optim as optim

# Training function
def train_model(model, train_loader, val_loader, epochs=10, lr=1e-4):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    # Loss & Optimizer
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr)

    for epoch in range(epochs):
        model.train()
        total_loss = 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()

            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()

        # Validation
        model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = torch.max(outputs, 1)
                correct += (predicted == labels).sum().item()
                total += labels.size(0)

        val_acc = 100 * correct / total
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {total_loss:.4f}, Val Accuracy: {val_acc:.2f}%")

# Example: Train ResNet-50
model = get_cnn_model("resnet50")
train_model(model, train_loader, val_loader, epochs=10, lr=3e-4)
```

```python
# Test function
def test_model(model, test_loader):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    model.eval()

    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            correct += (predicted == labels).sum().item()
            total += labels.size(0)

    test_acc = 100 * correct / total
    print(f"Test Accuracy: {test_acc:.2f}%")

# Example: Test ResNet-50
test_model(model, test_loader)
```

```python
# Train CNN Models
cnn_models = ["resnet50", "efficientnet", "densenet"]
for model_name in cnn_models:
    model = get_cnn_model(model_name)
    print(f"\nTraining {model_name}...")
    train_model(model, train_loader, val_loader)
    test_model(model, test_loader)

# Train Transformer Models
transformer_models = ["vit", "swin"]
for model_name in transformer_models:
    model = get_transformer_model(model_name)
    print(f"\nTraining {model_name}...")
    train_model(model, train_loader, val_loader)
    test_model(model, test_loader)

# Train Hybrid Models
hybrid_models = ["convnext", "cvt"]
for model_name in hybrid_models:
    model = get_hybrid_model(model_name)
    print(f"\nTraining {model_name}...")
    train_model(model, train_loader, val_loader)
    test_model(model, test_loader)
```

```python
import os
import torch
import kagglehub
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
from PIL import Image
from transformers import AutoImageProcessor, AutoModelForImageClassification

# ✅ Download dataset
dataset_path = kagglehub.dataset_download("ashenafifasilkebede/dataset")

# ✅ Load Hugging Face's CvT-13 model
processor = AutoImageProcessor.from_pretrained("microsoft/cvt-13")

model = AutoModelForImageClassification.from_pretrained(
    "microsoft/cvt-13",
    num_labels=2,  # 2 classes: Normal & OSCC
    ignore_mismatched_sizes=True  # ✅ Fix classifier mismatch issue
)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# ✅ Custom Dataset Class
class OSCCDataset(Dataset):
    def __init__(self, root_dir, processor):
        self.root_dir = root_dir
        self.processor = processor
        self.image_paths = []
        self.labels = []

        for label, class_name in enumerate(os.listdir(root_dir)):
            class_path = os.path.join(root_dir, class_name)
            if os.path.isdir(class_path):
                for img_name in os.listdir(class_path):
                    img_path = os.path.join(class_path, img_name)
                    self.image_paths.append(img_path)
                    self.labels.append(label)

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        img_path = self.image_paths[idx]
        label = self.labels[idx]

        image = Image.open(img_path).convert("RGB")
        image = self.processor(image, return_tensors="pt")["pixel_values"].squeeze(0)

        return image, torch.tensor(label)

# ✅ Load datasets
train_dataset = OSCCDataset(root_dir=os.path.join(dataset_path, "train"), processor=processor)
val_dataset = OSCCDataset(root_dir=os.path.join(dataset_path, "val"), processor=processor)
test_dataset = OSCCDataset(root_dir=os.path.join(dataset_path, "test"), processor=processor)

# ✅ Create DataLoaders
train_loader = DataLoader(train_dataset, batch_size=16, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=16, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=16, shuffle=False)

# ✅ Training Function
def train_model(model, train_loader, val_loader, epochs=10, lr=3e-4):
    model.train()
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=lr)

    for epoch in range(epochs):
        total_loss = 0
        correct, total = 0, 0

        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()

            outputs = model(images).logits
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            correct += (outputs.argmax(1) == labels).sum().item()
            total += labels.size(0)

        train_acc = 100 * correct / total
        val_acc = evaluate_model(model, val_loader)
        print(f"Epoch [{epoch+1}/{epochs}] - Loss: {total_loss:.4f}, Train Acc: {train_acc:.2f}%, Val Acc: {val_acc:.2f}%")

# ✅ Evaluation Function
def evaluate_model(model, loader):
    model.eval()
    correct, total = 0, 0

    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images).logits
            correct += (outputs.argmax(1) == labels).sum().item()
            total += labels.size(0)

    return 100 * correct / total

# ✅ Train CvT-13 Model
train_model(model, train_loader, val_loader, epochs=10, lr=3e-4)

# ✅ Test Model
test_acc = evaluate_model(model, test_loader)
print(f"Test Accuracy: {test_acc:.2f}%")
```
